## Trees Database - Ryan

```python
import pandas as pd
import numpy as np
import geopandas
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("trees.csv",low_memory=False)
```

```python
trees_benefit = data.groupby("neighborhood").sum()['air_quality_benfits_total_lbs'] # extract the data
trees_benefit_max = max(data.groupby("neighborhood").sum()['air_quality_benfits_total_lbs']) # Take only the necessary data
trees_benefit = trees_benefit/trees_benefit_max*100 # Make it on a scale of 1-100
trees_benefit.sort_values(ascending=False).head(10) # Show the ten "best" neighborhoods
```

```python
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp") # read in the shapefile
trees_map = neighborhoods.merge(trees_benefit, how='left', left_on='hood', right_on='neighborhood') # make the map
trees_map.plot(column='air_quality_benfits_total_lbs', # set the data to be used for coloring
               cmap='Greens',              # choose a color palette
               edgecolor="Black",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "Air Quality Benefit from Trees"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
plt.style.use('dark_background')
plt.show()
```

### Best neighborhood by trees: 
Squirrel Hill South does the best according to this metric.

## Parks Database - Shayaan

I will be using park data to find out where the best neighborhood in Pittsburgh for greenspace may reside.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

data = pd.read_csv("park.csv")
data.head()
```

```python
division_counts = data['divname'].value_counts().sort_index().sort_values(ascending=False)

division_counts.plot(kind='bar', color='skyblue')

plt.title('Number of Parks per Division', fontsize=16)
plt.xlabel('Division', fontsize=14)
plt.ylabel('Number of Parks', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

### Initial findings
The graph shows can show us the number of parks in each division. It appears as though Riverview has the most parks by a decent amount. It also appears that there is an outlier: Northeast. So lets drop it to make the graph look a little nicer.

```python
division_counts_filter = division_counts.drop('Northeast')
division_counts_filter.plot(kind='bar', color='skyblue')
plt.show()
```

That looks much better. It seems as though the Riverview division has the most parks, but the question is by how much. ### More Data Needed
So now we know that the Riverview division has the most parks located in it, but there is a question that needs to be asked: "Does Riverview have a lot of really tiny parks and Frick have a few large parks?" To answer this question it is relatively simple. Simply create a graph of the divisions sorted by total square feet of park.

```python
neighborhood_park_area = data.groupby('divname')['sqft'].sum()

neighborhood_park_area.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Total Square Feet of Parks by Neighborhood', fontsize=16)
plt.xlabel('Division', fontsize=12)
plt.ylabel('Total Square Feet', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Again it seems Northeast doesnt have much going for it, so lets drop it:

```python
data_filtered = data[data['divname'] != 'Northeast']

new_park_area = data_filtered.groupby('divname')['sqft'].sum().sort_values(ascending=False)

new_park_area.plot(kind='bar', color='skyblue')
plt.title('Total Square Feet per Division', fontsize=16)
plt.xlabel('Total Square Feet', fontsize=14)
plt.ylabel('Division Name', fontsize=14)
plt.tight_layout()
plt.show()
```

It seems as though MicKinley has the most total square feet of park in its division. In second place came Frick, which had the least amount of total parks. So it is a good thing that we looked at total square feet of park, and not just amount of parks. Also it is notable to see that Riverview was also second to last in total square feet of park even though it had the most parks. # Playing Fields Database - Nathan

I decided to do my part of the project on playing fields. specifically what neighboorhood as the most playing fields. My dataset consists of neighboorhoods, the names of playing fields, and which ones have fire and police zones. So I thought to create a way to filter the amount of playing fields in specific neighhborhoods by merging the data.

## Metric:

Chose to playing fields because it is a good place to play sports
My data set has all of the playing fields per neighborhood, the number of police zones/fire zone,s and lights.
For my metric, I went by how many playing fields there are per neighborhood

https://data.wprdc.org/dataset/playing-fields

First I import the needed tools to utilize geopandas and other graphing.

```python
import pandas as pd #importing file
import numpy as np
import geopandas 

%matplotlib inline 
import matplotlib.pyplot as plt

playF = pd.read_csv("PFS.csv")
```

Second, I need to count each of the neighboorhoods and the playing fields and then merge the data, and calculating how many there are per neighboorhood.

Third, I mapped it out using the .plot which utilizes the geopandas

```python
quantities = {}
for index, row in playF.iterrows():
    if row['neighborhood'] in quantities: #finding 
        quantities[row['neighborhood']] += 1
    else:
        quantities[row['neighborhood']] = 1

quantities = pd.DataFrame(list(quantities.items()), columns=['neighborhood', 'count']) #counting the neighborhoods
quantities_max = quantities['count'].max()

quantities['count'] = quantities['count'] / quantities_max #calculating the average
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
fields_map = neighborhoods.merge(quantities, how='left', left_on='hood', right_on='neighborhood') #merging the neighborhoods
fields_map.plot(column='count',
               cmap='Greens', 
               edgecolor = "Black",
               legend=True,             
               legend_kwds={'label': "number of fields"}, #changing the color
               figsize=(15, 10),        
               missing_kwds={"color": "lightgrey"} 
              )
plt.style.use('dark_background')
plt.show()
quantities.sort_values('count',ascending=False).head(10)
```

## Best Neighborhood:

According to the map, Brighton Heights has the most playing fields. The best neighborhood overall for green space would be Highland Park/East Liberty, which in my dataset has 4 playing fields. The chart above shows the density of the fields per neighborhood, with 1 set at most.

# Green Spaces - David

<h2>Introduction:</h2>

For my project in determining the best neighborhood, I chose green spaces. Green spaces have been proven to raise mental health as they increase the quality of life, and encourage positive mental health practices such as exercise and going for walks. Therefore the greenspaces data set was appropriate to measure this.

<h2>Metric:</h2>

For the metric, I chose the total acreage per neighborhood, I thought this was an appropriate metric to use since this measures the total amount of "green" per neighborhood.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

# loading in the data
green_spaces = pd.read_csv("green-spaces.csv")

# selecing the neighborhoods
neighborhoods = green_spaces['Neighborhood'].unique()

# Calculate total green space acreage per neighborhood
total_green_space = green_spaces.groupby('Neighborhood')['Total Acreage'].sum()

# Display in a bar graph
plt.figure(figsize=(10, 6))
total_green_space.plot(kind='bar', color='green')
plt.title('Total Green Space Acreage per Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Total Acreage')
plt.xticks(rotation=70)
plt.show()

total_green_space.sort_values(ascending=False).head(10)
```

<h2>Best Neighborhood:</h2>

As you can see the overall best neighborhood is East Liberty as it contains the most green space by acreage. This is due to the fact that it contains two of the the largest green spaces in Pittsburgh, when most neighborhoods only contain one if that with less acreage.
